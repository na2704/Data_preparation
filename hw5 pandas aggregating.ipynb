Load dataset nyc_weather_2018 and weather_stations, then calculate the average elevation of stations that calculat‘SNOW’ and contains ‘NJ US’ in its name.

```python
station.rename(columns={'id': 'station'}, inplace=True)
station.columns
```

```python
weather_snow=nyc[nyc.datatype=='SNOW']
weather_snow.head()
```

```python
station_us= station[station.name.str.contains('NJ US')]
station.head()
```

```python
info= weather_snow.merge(station_us, on='station')
info.shape
```

```python
info['station'].unique().shape
```

```python
info.drop_duplicates('station', inplace= True)
info.shape
```

```python
info[(info.datatype=='SNOW')&(info.name.str.contains('NJ US'))]['elevation'].mean()
```

Load the fb dataset, then:
- Calculate the percentage change between that and and last day in volumn (pct_change method)
- Sort the dataframe by percentage change then get five days that have the biggest change. 

```python
fb=pd.read_csv('fb_2018.csv')
fb
```

```python
fb.volume.pct_change
```

```python
nfb=fb.assign(pct_change= lambda x: x.volume\
             .pct_change())
nfb
```

```python

nfb.sort_values(by = 'pct_change', ascending= False).head(5)
```

```python
dt1 = pd.read_csv('earthquakes.csv')
dt1 = dt1[(dt1['magType'] == 'ml')]
dt1.head()
```

#Create bins for each full number of earthquake magnitude (for instance, the first bin is (0, 1], the second is (1, 2], and so on) with the ml magnitude type and count how many are in each bin.

```python
mag_binned = pd.cut(dt1.mag, [0,1,2,3,4,5], \
                    labels= ['low', 'normal', 'medium', 'high','dangeous'])
mag_binned.value_counts()
```
